Data Cleaning and Standardizing

```python
matches.columns = matches.columns.str.lower()
matches.columns = matches.columns.str.replace(' ', '_')
```

```python
matches['city']= matches['city'].fillna('unknown')
```

```python
matches['method'].value_counts()
```

```python
matches['method']= matches['method'].fillna('No DLS')
```

```python
matches['date'] = pd.to_datetime(matches['date'])
```

```python
matches.drop_duplicates(inplace=True)
```

```python
matches.isnull().sum()
```

```python
matches.duplicated().sum()
```

```python
deliveries.columns = deliveries.columns.str.lower()
deliveries.columns = deliveries.columns.str.replace(' ', '_')
```
